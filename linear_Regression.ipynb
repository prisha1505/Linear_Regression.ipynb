```python
# # Data Preprocessing
# Importing the libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Importing the dataset
url="https://raw.githubusercontent.com/singhudaypratap/ML_LAB_datasets/main/Salary_Data.csv"
dataset = pd.read_csv(url)
dataset
```

```python
dataset.head()
```

```python
dataset.info()
```

```python
dataset.describe().T
```

```python
dataset.isnull().sum
```

```python
# Data visualization
from matplotlib import pyplot
dataset.hist()
pyplot.show()
```

```python
# Data visualization
dataset.plot(kind='density' ,subplots=True, layout=(3,3), sharex=False)
pyplot.show()
```

```python
# Extracting dependent and independent variables:
# Extracting independent variable:
X = dataset.iloc[:, 0:1].values
# Extracting dependent variable:
y = dataset.iloc[:, 1].values # this is for 1 d array  coming up this will make it 2-d array
```

```python
print(X)
print(y)
```

```python
print(y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 1/3, random_state = 0)
```

```python
print(X_train.shape)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train,y_train)
```

```python
y_pred=lr.predict(X_test)
```

```python
# Visualising the Training set results 5
plt.scatter(X_train, y_train, color = 'red')
plt.plot(X_train, lr.predict(X_train), color = 'blue')
plt.title('Salary vs Experience (Training set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

```python
# Visualising the Test set results
plt.scatter(X_test, y_test, color = 'red')
plt.plot(X_test, lr.predict(X_test), color = 'blue')
plt.title('Salary vs Experience (Test set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

```python
print("Regressor slope:  %.2f  "%( lr.coef_[0]))
print("Regressor intercept:%.2f  "% lr.intercept_)
```

```python
YearsExperience= 6
print("Salary for given Years of Experience is :     %2f"    %(lr.predict([[YearsExperience]])))
```

```python
from sklearn import metrics
print("RMSE %2.f" %(np.sqrt(metrics.mean_absolute_error(y_test,y_pred))))
print('Train Score: %f' %(lr.score(X_train, y_train)))
print('Test Score: %f' % (lr.score(X_test, y_test)) )
print("Underfitting")
```
